```python
from sklearn.preprocessing import RobustScaler
WEIGHTS = {
    "systolic_bp": 0.15,
    "diastolic_bp": 0.10,
    "heart_rate": 0.10,
    "bmi": 0.05,
    "glucose": 0.20,
    "hba1c": 0.30,
    "med_adherence": -0.25,
}

DELTA_THRESHOLD = 0.15
ABS_THRESHOLD = 0.25
LOOKAHEAD_DAYS = 90
```

```python
def compute_composite_score(df, weights=WEIGHTS):
    """Return df with normalized columns and a composite score column 'health_score'."""
    df = df.copy()
    numeric_cols = [c for c in weights.keys()]
    scaler = RobustScaler()
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    denom = sum(abs(w) for w in weights.values())
    df['health_score'] = sum(df[col] * w for col, w in weights.items()) / denom
    return df, scaler
```

```python
def create_90d_label(df, lookahead=LOOKAHEAD_DAYS,
                     delta_thresh=DELTA_THRESHOLD, abs_thresh=ABS_THRESHOLD):
    """
    For each patient and each day t, find score at closest timestamp >= t+lookahead.
    If score increases beyond threshold (relative or absolute) label=1 else 0.
    Returns a DataFrame of (patient_id, day, health_score_t, health_score_t90, y)
    """
    df = df.copy()
    # ensure day numeric/sorted
    df = df.sort_values(['patient_id', 'day']).reset_index(drop=True)
    records = []
    # convert days to numeric if needed
    for pid, group in df.groupby('patient_id'):
        days = group['day'].values
        scores = group['health_score'].values
        # use searchsorted for efficiency
        for i, t in enumerate(days):
            t90 = t + lookahead
            # find first index with day >= t90
            j_idx = np.searchsorted(days, t90, side='left')
            if j_idx >= len(days):
                # no follow up >= t+90 -> skip or label unknown
                continue
            score_t = scores[i]
            score_t90 = scores[j_idx]
            # relative increase
            rel_inc = (score_t90 - score_t) / (abs(score_t) + 1e-6)
            abs_inc = score_t90 - score_t
            y = int((rel_inc >= delta_thresh) or (abs_inc >= abs_thresh))
            records.append({
                'patient_id': pid,
                'day': t,
                'score_t': score_t,
                'score_t90': score_t90,
                'rel_inc': rel_inc,
                'abs_inc': abs_inc,
                'y': y
            })
    lab_df = pd.DataFrame.from_records(records)
    return lab_df
```

```python

SEQ_FEATS = ['systolic_bp','diastolic_bp','heart_rate','bmi','glucose','hba1c','med_adherence']
STATIC_FEATS = ['age','sex','diabetes','hypertension','heart_disease','obesity','copd','asthma','ckd','depression','hyperlipidemia']
WINDOW_SIZE = 90  # days

def build_dataset_windows(df, labels_df, window_size=WINDOW_SIZE, seq_feats=SEQ_FEATS, static_feats=STATIC_FEATS):
    """
    Returns:
      X: np.array shape (N, window_size, n_feats)
      mask: np.array shape (N, window_size) with 1 for real timesteps, 0 for padded
      static: np.array shape (N, len(static_feats))
      y: np.array shape (N,)
      meta_df: pandas dataframe with patient_id, day
    """
    # merge labels to df
    df = df.sort_values(['patient_id','day']).reset_index(drop=True)
    lab = labels_df[['patient_id','day','y']].copy()
    merged = df.merge(lab, on=['patient_id','day'], how='left')
    # Only keep rows with label not null (training examples)
    merged = merged[merged['y'].notnull()].reset_index(drop=True)

    X_list, mask_list, static_list, y_list, meta = [], [], [], [], []
    # we'll need full groups for indexing days -> easier to iterate per patient from original df
    grouped = df.groupby('patient_id')
    for idx, row in merged.iterrows():
        pid = row['patient_id']
        day_t = row['day']
        group = grouped.get_group(pid).reset_index(drop=True)
        # locate index of current day in full group
        pos = int(np.where(group['day'].values == day_t)[0][0])
        start_idx = max(0, pos - window_size + 1)
        seq_rows = group.loc[start_idx:pos, seq_feats]
        seq_len = len(seq_rows)
        # pad if needed (left pad)
        if seq_len < window_size:
            pad_len = window_size - seq_len
            pad = np.zeros((pad_len, len(seq_feats)), dtype=float)
            seq_arr = np.vstack([pad, seq_rows.values])
            mask = np.array([0]*pad_len + [1]*seq_len, dtype=int)
        else:
            seq_arr = seq_rows.values[-window_size:]
            mask = np.ones(window_size, dtype=int)
        # static: take last known static (row values)
        static_vals = row[static_feats].values.astype(float)
        X_list.append(seq_arr.astype(float))
        mask_list.append(mask)
        static_list.append(static_vals)
        y_list.append(int(row['y']))
        meta.append({'patient_id': pid, 'day': day_t})
    X = np.stack(X_list)           # (N, W, F)
    mask = np.stack(mask_list)     # (N, W)
    static_arr = np.stack(static_list)  # (N, n_static)
    y = np.array(y_list)
    meta_df = pd.DataFrame(meta)
    return X, mask, static_arr, y, meta_df
```

```python
import torch
from torch.utils.data import Dataset

class SeqHealthDataset(Dataset):
    def __init__(self, X, mask, static, y):
        self.X = torch.tensor(X, dtype=torch.float32)            # (N, W, F)
        self.mask = torch.tensor(mask, dtype=torch.float32)      # (N, W)
        self.static = torch.tensor(static, dtype=torch.float32)  # (N, S)
        self.y = torch.tensor(y, dtype=torch.float32)            # (N,)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.mask[idx], self.static[idx], self.y[idx]
```

```python
# model.py
import torch
import torch.nn as nn
import torch.nn.functional as F

class LSTMClassifier(nn.Module):
    def __init__(self, seq_feat_dim, static_dim, hidden_size=128, num_layers=1, dropout=0.3, bidirectional=False):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_directions = 2 if bidirectional else 1
        self.lstm = nn.LSTM(
            input_size=seq_feat_dim,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=bidirectional,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(hidden_size * self.num_directions + static_dim, 64)
        self.fc2 = nn.Linear(64, 1)  # final logit output

    def forward(self, x, mask, static_feats):
        lengths = mask.sum(dim=1).long().cpu()
        packed = nn.utils.rnn.pack_padded_sequence(x, lengths, batch_first=True, enforce_sorted=False)
        packed_out, (hn, cn) = self.lstm(packed)

        if self.num_directions == 1:
            last_hidden = hn[-1]  # (B, hidden)
        else:
            last_hidden = torch.cat([hn[-2], hn[-1]], dim=1)  # (B, hidden*2)

        h = self.dropout(last_hidden)
        h = torch.cat([h, static_feats], dim=1)
        h = F.relu(self.fc1(h))
        logits = self.fc2(h).squeeze(1)  # raw logits (no sigmoid)
        return logits
```

```python
# train_eval.py
import torch
import numpy as np
from sklearn.metrics import roc_auc_score, average_precision_score, precision_score, recall_score, f1_score
from torch.utils.data import DataLoader

def compute_metrics(y_true, y_scores, threshold=0.5):
    y_pred = (y_scores >= threshold).astype(int)
    metrics = {
        'auroc': float(roc_auc_score(y_true, y_scores)) if len(np.unique(y_true)) > 1 else None,
        'auprc': float(average_precision_score(y_true, y_scores)),
        'precision': float(precision_score(y_true, y_pred, zero_division=0)),
        'recall': float(recall_score(y_true, y_pred, zero_division=0)),
        'f1': float(f1_score(y_true, y_pred, zero_division=0))
    }
    return metrics

def train_model(model, train_dataset, val_dataset, device,
                batch_size=64, epochs=30, lr=1e-3, pos_weight=None, patience=5):
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    model.to(device)
    if pos_weight is not None:
        pos_w = torch.tensor([pos_weight], dtype=torch.float32).to(device)
        criterion = torch.nn.BCEWithLogitsLoss(pos_weight=pos_w)
    else:
        criterion = torch.nn.BCEWithLogitsLoss()

    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    best_val_auprc = -np.inf
    best_state = None
    epochs_no_improve = 0

    for epoch in range(1, epochs + 1):
        model.train()
        total_loss = 0.0
        for X, mask, static, y in train_loader:
            X = X.to(device); mask = mask.to(device); static = static.to(device); y = y.to(device)
            optimizer.zero_grad()
            logits = model(X, mask, static)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * y.size(0)
        train_loss = total_loss / len(train_dataset)

        # Validation
        model.eval()
        y_true, y_scores = [], []
        with torch.no_grad():
            for X, mask, static, y in val_loader:
                X = X.to(device); mask = mask.to(device); static = static.to(device)
                logits = model(X, mask, static)
                probs = torch.sigmoid(logits).cpu().numpy()
                y_scores.extend(probs.tolist())
                y_true.extend(y.numpy().tolist())
        y_true = np.array(y_true); y_scores = np.array(y_scores)
        val_metrics = compute_metrics(y_true, y_scores)
        val_auprc = val_metrics['auprc']

        print(f"Epoch {epoch} train_loss={train_loss:.4f} val_auprc={val_auprc:.4f} val_auroc={val_metrics['auroc']}")
        if val_auprc > best_val_auprc:
            best_val_auprc = val_auprc
            best_state = model.state_dict()
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping.")
                break
    if best_state is not None:
        model.load_state_dict(best_state)
    return model, best_val_auprc, val_metrics
```

```python
import torch
from torch.utils.data import random_split

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
full_dataset = SeqHealthDataset(X, mask, static_arr, y)
```

```python
from torch.utils.data import random_split

train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

print(f"Train size: {len(train_dataset)}, Validation size: {len(val_dataset)}")
```

```python
seq_feat_dim = X.shape[2]     # features per day
static_dim = static_arr.shape[1]  # static features

model = LSTMClassifier(
    seq_feat_dim=seq_feat_dim,
    static_dim=static_dim,
    hidden_size=128,
    num_layers=1,
    dropout=0.3,
    bidirectional=True
)
```

```python
model, best_val_auprc, val_metrics = train_model(
    model,
    train_dataset,
    val_dataset,
    device=device,
    batch_size=64,
    epochs=50,
    lr=1e-3,
    patience=20
)

print("Best validation AUPRC:", best_val_auprc)
print("Validation metrics:", val_metrics)
```
